## LIMPANDO OS DADOS

```python
#Limpando a lista de forma de pagamento
columns_drop = ['payment_sequential']
df_forma_de_pagamento = df_forma_de_pagamento.drop(columns_drop, axis=1)
df_forma_de_pagamento
```

```python
#Tirando informações que não serão necessárias da lista de pedidos
columns_drop = ['order_item_id', 'shipping_limit_date']
df_pedidos = df_pedidos.drop(columns_drop, axis=1)
df_pedidos
```

```python
#Limpando dados da lista de produtos
#Queria tirar mais colunas mas deu erro diversas vezes, então ficou dessa forma
columns_drop = ['product_name_lenght']
df_produtos = df_produtos.drop(columns_drop, axis=1)
df_produtos
```

```python
columns_drop = ['seller_zip_code_prefix']
df_vendedores= df_vendedores.drop(columns_drop, axis=1)
df_vendedores
```

## Juntando os DataSet

```python
#Df_pedidos + df_vendedores
df_olist1 = pd.merge(df_pedidos, df_vendedores, on='seller_id')
df_olist1
```

```python
# mais df_produtos
df_olist_f1 = pd.merge(df_olist1, df_produtos, on='product_id')
df_olist_f1
```

```python
delete_columns = [ 'product_description_lenght', 'product_photos_qty','product_length_cm','product_height_cm','product_width_cm']
df_olist_f1= df_olist_f1.drop(delete_columns, axis=1)
df_olist_f1
```

```python
df_olist_f1.to_csv('base_olist_f1.csv', index=False)
```
